```python
# Ορίζουμε το αρχικό url του site
main_url = 'https://www.newsit.gr/eidhseis-tora/'
# Ξεκινάμε με μια άδεια λίστα
stories_list = []
stories = doc.find('div', {'class': 'inside-article'}).find_all('article')
for story in stories:
    # Φτιάχνουμε ένα λεξικό χωρίς τίποτα
    story_dict = {}
    # link
    try:
      link = story.find('h2', {'class': 'full-link'})['href'] 
      # ελέγχουμε αν το link ξεκινάει από "http"
      if link.startswith("http"):
        story_dict['url'] = link
      else:
        story_dict['url'] = main_url + link
    except:
      story_dict['url'] = ''
      pass
    # title
    try:
      title = story.find('h2', {'class': 'mb-1'}).text
      story_dict['title'] = title
    except:
      story_dict['title'] = ''
      pass
    # time
    try:
      time = story.find('time').text
      story_dict['time'] = time
    except:
      story_dict['time'] = ''
      pass
    # section
    try:
      section = story.find('a', {'class': 'grid-container.grid-parent'}).text
      story_dict['section'] = section
    except:
      story_dict['section'] = ''
      pass
    # summary
    try:
      summary = story.find('div', {'class': 'mb-1'}).text
      story_dict['summary'] = summary
    except:
      story_dict['summary'] = ''
      pass
    # Προσθέτουμε το λεξικό στη λίστα 
    stories_list.append(story_dict)
    
# Δημιουργούμε dataframe από τη λίστα με τα dictionaries
newsit_df = pd.DataFrame(stories_list)
```

```python
# Ορίζουμε το αρχικό url του site
main_url = 'https://www.newsit.gr/eidhseis-tora/'
# Ορίζουμε το βασικό url για τις σελίδες που θέλουμε
start_url = 'https://www.newsit.gr/eidhseis-tora/page/2/3/4'
# Φτιάχνουμε μια κενή λίστα για να αποθηκεύουμε τα λεξικά που θα φτιάχνουμε για κάθε άρθρο
stories_list = []

# φτιάχνουμε τη λούπα που ο scraper θα μπαίνει σε καθε μια από τις σελίδες για να φέρνει τα άρθρα
for page_num in range(0, 2):
  url = start_url + str(page_num)
  response = requests.get(url)
  doc = BeautifulSoup(response.text, 'html.parser')
  stories = doc.find('div', {'class': 'inside-article'}).find_all('article')
  for story in stories:
      # Φτιάχνουμε ένα λεξικό χωρίς τίποτα
      story_dict = {}
      # link
      try:
        link = story.find('a', {'class': 'full-link'})['href'] 
        # ελέγχουμε αν το link ξεκινάει από "http"
        if link.startswith("http"):
          story_dict['url'] = link
        else:
          story_dict['url'] = main_url + link
      except:
        story_dict['url'] = ''
        pass
      # title
      try:
        title = story.find('a', {'class': 'mb-1'}).text
        story_dict['title'] = title
      except:
        story_dict['title'] = ''
        pass
      # time
      try:
        time = story.find('time').text
        story_dict['time'] = time
      except:
        story_dict['time'] = ''
        pass
      # section
      try:
        section = story.find('a', {'class': 'inside-article__category'}).text
        story_dict['section'] = section
      except:
        story_dict['section'] = ''
        pass
      # summary
      try:
        summary = story.find('div', {'class': 'inside-article__summary'}).text
        story_dict['summary'] = summary
      except:
        story_dict['summary'] = ''
        pass
      # Προσθέτουμε το λεξικό στη λίστα 
      stories_list.append(story_dict)
    
# Δημιουργούμε dataframe από τη λίστα με τα dictionaries
newsIt_df = pd.DataFrame(stories_list)
```
